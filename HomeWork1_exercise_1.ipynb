```python
X1 = data.iloc[:,0]
X2 = data.iloc[:,1]
Y = data.iloc[:,2]
plt.scatter(X1,Y,color = 'black')
plt.scatter(X2,Y,color = 'red')
```

```python
theta0_list = []
theta1_list = []
theta2_list = []
cost_list = []

theta0 = 0.0
theta1 = 0.01
theta2 = 0.02
epochs = 800000
alpha = 0.0001
#the formula is
#Y_pred = theta0 + theta1*X1 + theta2*X2
Y_pred= theta0 + theta1*X1 + theta2*X2
```

```python
m = float(len(X1))
iteration = []
#performing gradient descent
print('theta0\t, theta1\t, cost', 'Epoch')
for i in range(epochs):
    if i<100:
        Y_pred = theta0 + theta1*X1 + theta2*X2
        temp0 = (2/m)*sum(Y_pred-Y)
        temp1 = (2/m)*sum(X1*(Y_pred-Y))
        temp2 = (2/m)*sum(X2*(Y_pred-Y))
        theta0 = theta0 - alpha*temp0
        theta1 = theta1 - alpha*temp1
        theta2 = theta2 - alpha*temp2
        theta0_list.append(theta0)
        theta1_list.append(theta1)
        theta2_list.append(theta2)

        Y_pred = theta0 + theta1*X1 + theta2*X2
        cost = np.sum((Y_pred - Y)**2)
        cost_list.append(cost)
        print(theta0,'\t',theta1,'\t',theta2,'\t', cost)
        plt.scatter(X1,Y)
        plt.plot([min(X1),max(X1)],[min(Y_pred), max(Y_pred)], color = 'red')
        plt.show()
        iteration.append(i)
        Y_pred = theta0 + theta1*X1 + theta2*X2
        temp0 = (2/m)*sum(Y_pred-Y)
        temp1 = (2/m)*sum(X1*(Y_pred-Y))
        temp2 = (2/m)*sum(X2*(Y_pred-Y))
        theta0 = theta0 - alpha*temp0
        theta1 = theta1 - alpha*temp1
        theta2 = theta2 - alpha*temp2
        theta0_list.append(theta0)
        theta1_list.append(theta1)
        theta2_list.append(theta2)

        Y_pred = theta0 + theta1*X1 + theta2*X2

        cost = np.sum((Y_pred - Y)**2)
        cost_list.append(cost)
        iteration.append(i)

print(theta0, '\t', theta1, '\t',theta2, '\t', cost)
plt.scatter(X1,Y)
plt.plot([min(X1),max(X1)],[min(Y_pred), max(Y_pred)], color = 'red')
```
